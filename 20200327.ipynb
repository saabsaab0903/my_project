```python
from bs4 import BeautifulSoup
import requests as re
import pandas as pd

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = re.get(url)
print(r) #200=ok
soup = BeautifulSoup(r.text, 'html.parser')
# class_ = title -> text 
f = open('nba_ptt.txt', 'w', encoding='utf8') #write
title = soup.find_all(class_ = 'title')
author = soup.find_all(class_ = 'author')
date = soup.find_all(class_ = 'date')
titles = []
authors = []
dates = []
for t,a,d in zip(title, author, date):
#     print(t.text.rstrip(), a.text, d.text)
    titles.append(t.text)
    authors.append(a.text)
    dates.append(d.text)
#     f.writelines(t.text)

for i in range(5):
#     print('page', i+1)
    next_page_btn = soup.find_all(class_="btn wide")
    url = 'https://www.ptt.cc/' + next_page_btn[1].get('href')
    r = re.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    title = soup.find_all(class_ = 'title')
    author = soup.find_all(class_ = 'author')
    date = soup.find_all(class_ = 'date')
    for t,a,d in zip(title, author, date):
#         print(t.text.rstrip(), a.text, d.text)
        titles.append(t.text)
        authors.append(a.text)
        dates.append(d.text)

dict = {
    'title':titles,
    'author':authors,
    'date':dates
}

df = pd.DataFrame(dict)
print(df.describe())
print(df.head(10))
print(df.tail(3))
print(df.shape)
df.to_csv('tt.csv')
f.close()
```
